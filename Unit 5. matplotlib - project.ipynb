# 5. 태어난 해부터의 생일 선택하기

```python
import csv
f = open('temp.csv')
data = csv.reader(f)
birth = input('생년월일을 입력해주세요(예:2000-01-01) :')
# 생일을 '-'을 기준으로 나눠서 birth 변수에 다시 저장(리스트로)
birth = birth.split('-')

for row in data :
    # 날짜를 '-'을 기준으로 나눠서 date 변수에 저장(리스트로)
    date = row[0].split('-')
    # 생년부터의 생일날 최고 기온 데이터를 저장하려면 조건을 생각해보세요.
    if int(date[0]) >= int(birth[0]) and date[1] == birth[1] and date[2] == birth[2] :
        print(row)
f.close()
```

# 6. 생일에 대한 최고, 최저 기온 데이터 저장하기(리스트에)

```python
import csv
f = open('temp.csv')
data = csv.reader(f)
birth = input('생년월일을 입력해주세요(예:2000-01-01) :')
# 생일을 '-'을 기준으로 나눠서 birth 변수에 다시 저장(리스트로)
birth = birth.split('-')

high = []

for row in data :
    # 날짜를 '-'을 기준으로 나눠서 date 변수에 저장
    date = row[0].split('-')
    # 생년부터의 생일날 최고 기온 데이터를 저장하려면 조건을 생각해보세요.
    if int(date[0]) >= int(birth[0]) and date[1] == birth[1] and date[2] == birth[2] :
        print(row)
        high.append(float(row[-1]))

print(high)
f.close()
```

```python
import csv
f = open('temp.csv')
data = csv.reader(f)
birth = input('생년월일을 입력해주세요(예:2000-01-01) :')
birth = birth.split('-')
# 같은 패턴으로 최저 기온 데이터를 low에 저장해보세요.
high = []
low = []

for row in data :
    date = row[0].split('-')
    if int(date[0]) >= int(birth[0]) and date[1] == birth[1] and date[2] == birth[2] :
        high.append(float(row[-1]))
        low.append(float(row[-2]))

print(high)
print(low)
f.close()
```

# 7. matplotlib 라이브러리 불러오기
# 8. 그래프로 표현하기

```python
import csv
f = open('temp.csv')
data = csv.reader(f)
birth = input('생년월일을 입력해주세요(예:2000-01-01) :')

birth = birth.split('-')
high = []
low = []

for row in data :
    date = row[0].split('-')
    if int(date[0]) >= int(birth[0]) and date[1] == birth[1] and date[2] == birth[2] :
        high.append(float(row[-1]))
        low.append(float(row[-2]))
f.close()

# high, low 리스트를 그래프로 출력합니다.
import matplotlib.pyplot as plt
plt.figure(dpi = 300)
plt.plot(high)
plt.plot(low)
plt.show()
```

# 9. 보기 좋게 꾸며주기

```python
import csv
f = open('temp.csv')
data = csv.reader(f)
birth = input('생년월일을 입력해주세요(예:2000-01-01) :')
birth = birth.split('-')
high = []
low = []

for row in data :
    date = row[0].split('-')
    if int(date[0]) >= int(birth[0]) and date[1] == birth[1] and date[2] == birth[2] :
        high.append(float(row[-1]))
        low.append(float(row[-2]))
f.close()

# label을 max, min으로 정해서 범례를 출력해주세요.
import matplotlib.pyplot as plt
plt.plot(high, label = 'max')
plt.plot(low, label = 'min')
plt.legend()
plt.show()
```

```python
import csv
f = open('temp.csv')
data = csv.reader(f)
birth = input('생년월일을 입력해주세요(예:2000-01-01) :')
birth = birth.split('-')
high = []
low = []

for row in data :
    date = row[0].split('-')
    if int(date[0]) >= int(birth[0]) and date[1] == birth[1] and date[2] == birth[2] :
        high.append(float(row[-1]))
        low.append(float(row[-2]))
        
f.close()
        
import matplotlib.pyplot as plt
plt.figure(dpi = 300)
plt.rc('font',family = 'Malgun Gothic') 
# plt.rc('font',family = 'Apple Gothic')
plt.plot(high, label = 'max')
plt.plot(low, label = 'min')
plt.rcParams['axes.unicode_minus'] = False
plt.title(birth[0] + '년부터의 ' + birth[1] + '월 '+ birth[2] + '일의 온도 변화 그래프')
plt.xticks(range(0,len(high),5), range(int(birth[0]),int(birth[0])+len(high)+1,5))
plt.legend()
plt.show()
```
